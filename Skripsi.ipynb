```python
def VizDist(df,var):
  plt.figure(figsize=(6, 4))

  # Histogram
  plt.subplot(2, 1, 1)
  sns.histplot(df[var], kde=True, color='skyblue')
  plt.xlabel('')
  plt.xticks([])

  # Boxplot
  plt.subplot(2, 1, 2)
  sns.boxplot(x=df[var], width = 0.4)
  plt.tight_layout()
  plt.suptitle('Distribution Variable ' + var, y=1.05, fontsize=14,fontweight='bold')
  plt.show()
```

# Transformation

```python
# To Visualize all Variable
def BoxplotAllVar(df, trans):
  plt.figure(figsize=(12, 8))
  df_num = df.select_dtypes(include='number')
  sns.boxplot(data = df_num, width = 0.5, palette = sns.color_palette("pastel"), orient='h')
  plt.title('Boxplot of All Variables with ' + trans)
  plt.xticks(rotation=90)
  plt.show()
```

## Min-Max Scaling

```python
def MinMaxScaling(df):
  # selecting variable numeric
  df_num = df.select_dtypes(include='number')

  scaler = MinMaxScaler()
  scaled_data = scaler.fit_transform(df_num)
  scaled_df = pd.DataFrame(scaled_data, columns=df_num.columns)
  data = df.copy()
  data[df_num.columns] = scaled_df
  return data

def MinMaxSlope(df_slope):
  slope_max = df_slope.max()
  slope_min = df_slope.min()

  result = (slope_max - df_slope)/(slope_max - slope_min)
  return(result)
```

```python
df_MinMax = MinMaxScaling(dataset)
df_MinMax['Slope'] = MinMaxSlope(dataset['Slope'])
df_MinMax.describe()
```

```python
BoxplotAllVar(df_MinMax, 'Min Max Scaling')
```

## Log Transformation

```python
def LogTransform(df):
  small_constant = 1e-5
  df_temp = df.select_dtypes(include='number')
  df_transformed = np.log1p(df_temp + small_constant)
  data = df.copy()
  data[df_transformed.columns] = df_transformed
  return data
```

```python
df_LogTransform = LogTransform(dataset)
df_LogTransform.describe()
```

```python
BoxplotAllVar(MinMaxScaling(df_LogTransform), 'Logarithmic')
```

## YeoJohnson Transformation

```python
from sklearn.preprocessing import PowerTransformer

def YJTransform(df):
  df_temp = df.select_dtypes(include='number')

  pt = PowerTransformer()
  pt.fit(df_temp)
  lambda_values = pt.lambdas_
  trans = pt.transform(df_temp)

  data = df.copy()
  data[df_temp.columns] = trans

  return data, lambda_values
```

```python
df_YJTransform, lambda_values = YJTransform(dataset)
df_YJTransform.describe()
```

```python
lambda_values
```

```python
BoxplotAllVar(df_YJTransform, 'Yeo Johnson')
```

```python
def CorPlot(df):
  df_temp = df.select_dtypes(include='number')
  correlation_matrix = df_temp.corr()

  plt.figure(figsize=(8, 6))
  sns.heatmap(correlation_matrix, annot=True, cmap = 'coolwarm', fmt=".2f", linewidths=.25, annot_kws={"size": 8})
  plt.title('Correlation Plot')
  plt.show()

  # print(correlation_matrix)
```

# Select Variables

```python
def SelectVar(df):
  aspek_ekonomi = df[['HSI', 'NTL', 'NDVI', 'NDBI', 'NDWI', 'Slope', 'elevation']]
  aspek_sosial = df[['NTL', 'NDVI', 'VHI', 'SAVI']]
  aspek_fisik = df[['HSI', 'NTL', 'NDVI', 'Adj_World_Pop', 'RWI']]
  return aspek_ekonomi, aspek_sosial, aspek_fisik
```

# Geometrical Mean

```python
from scipy.stats import gmean
def GeomMean(df, constant_value):
  df_add = df+constant_value
  geometric_means = df_add.apply(gmean, axis = 1)
  return geometric_means

def GeomAllVar(df, constant_value):
  # selecting variable every Aspect
  df_eko, df_sos, df_fis = SelectVar(df)

  # Aspek Ekonomi
  df_eko['Geom_Mean'] = GeomMean(df_eko, constant_value)

  # Aspek Sosial
  df_sos['Geom_Mean'] = GeomMean(df_sos, constant_value)

  # Aspek Fisik
  df_fis['Geom_Mean'] = GeomMean(df_fis, constant_value)

  return df_eko, df_sos, df_fis
```

### Min-Max Scale

```python
G_MinMax_result_eko, G_MinMax_result_sos, G_MinMax_result_fis = GeomAllVar(df_MinMax, 1)
```

### Log Transform

```python
G_Log_result_eko, G_Log_result_sos, G_Log_result_fis = GeomAllVar(df_LogTransform, 6)
```

## Yeo-Johnson Transformation

```python
G_YJ_result_eko, G_YJ_result_sos, G_YJ_result_fis = GeomAllVar(df_YJTransform, 4)
```

# Harmonical Means

```python
from scipy.stats import hmean
def HarMean(df, constant_value):
  df_add = df+constant_value
  harmonical_means = df_add.apply(hmean, axis = 1)
  return harmonical_means

def HarAllVar(df, constant_value):
  # selecting variable every Aspect
  df_eko, df_sos, df_fis = SelectVar(df)

  # Aspek Ekonomi
  df_eko['Har_Mean'] = HarMean(df_eko, constant_value)

  # Aspek Sosial
  df_sos['Har_Mean'] = HarMean(df_sos, constant_value)

  # Aspek Fisik
  df_fis['Har_Mean'] = HarMean(df_fis, constant_value)

  return df_eko, df_sos, df_fis
```

## Min Max

```python
H_MinMax_result_eko, H_MinMax_result_sos, H_MinMax_result_fis = HarAllVar(df_MinMax,1)
```

## Log Transform

```python
H_Log_result_eko, H_Log_result_sos, H_Log_result_fis = HarAllVar(df_LogTransform, 6)
```

## Yeo-Johnson

```python
H_YJ_result_eko, H_YJ_result_sos, H_YJ_result_fis = HarAllVar(df_YJTransform, 4)
```
